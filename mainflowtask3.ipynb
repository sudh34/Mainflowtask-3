```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv('C:\\Users\\sudha\\Downloads\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#scatter plot with year against own 
plt.scatter(data['year'],data['own'])

#adding the title to the plot 
plt.title('scatter plot ')

plt.xlabel('year')
plt.ylabel('own')

#showing the plot 

plt.show()
```

```python
#line chart of year against own 

plt.plot(data['year'])
plt.plot(data['own'])

plt.title('line chart')

plt.xlabel(' year')
plt.ylabel('own')

plt.show()
```

```python
#barchart or barplot 

plt.bar(data['year'],data['own'])

plt.title('barchart')

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#histogram 

plt.hist(data['income'])

plt.title('histogram ')

plt.show()
```
